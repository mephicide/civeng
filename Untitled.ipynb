```python
distributedLoad = {
    'begin' : 0,
    'end' : 1,
    'live' : 10,
    'roofLive' : 3,
    'dead' : 4,
    'snow' : 1,
    'earth' : 0,
    'seismic' : 20,
    'wind' : 13
}
```

```python
def distributeLoad(start, end, load, beamLength):
    totalShear = load * (end - start)
    totalShearLoc = (end-start) / 2
    rightReaction = totalShear * totalShearLoc / beamLength
    leftReaction = totalShear - rightReaction
    return {'left': leftReaction, 'right':rightReaction}
```

```python
beamLength=15
distributeLoad(distributedLoad['begin'], distributedLoad['end'], distributedLoad['wind'], beamLength)
```

```python
for loadType in [k for k in distributedLoad.keys() if k !='end' and k!='begin']:
    print(distributeLoad(distributedLoad['begin'], distributedLoad['end'], distributedLoad[loadType], beamLength))
```
